## Espacio Muestral.

Si se realiza un experimento, un resultado particular de ese experimento se lo denomina **muestra puntual**,  $\zeta$. El conjunto de todos los resultados obtenibles en un experimento es el **espacio muestral** $ S=\left\{\zeta _i\right\}$  con $i \in \mathbb{N}$. 

Como ejemplo, de **espacio muestral discreto**, planteamos el experimento de arrojar una moneda al aire tres veces y anotamos la secuencia que devuelve, cara: $C$ o seca: $S$. De tal manera, la muestra número 40 del experimento es $\zeta^1 _{40}=CSS$ siendo el espacio muestral:

$$S_1=\left\{CCC, CCS, CSC, CSS, SCC, SCS, SSC, SSS\right\}$$ 



## Eventos

Un evento es un subconjunto del espacio muestral $S$. Se puede definir al evento $A_1$: **dos caras de tres** , en otros términos:

$$A_1=\left\{CCS, SCC, CSC  \right\}$$ 

El evento así definido, se puede observar mediante el estudio de un conjunto de muestras adquiridas en un experimento apropiado, en ese sentido, podemos decir que "sucedió tal evento porque así lo reflejó una muestra". 


## Medida de Probabilidad.

Se define una $\sigma-$álgebra $\mathcal{F}$ sobre $S$. Es decir $\mathcal{F}$ es el conjunto de todos los posibles eventos formados por relizaciones de $\zeta \in S$.


Una medida de probabilidad $P$ sobre $\mathcal{F}$ es una medida que asigna a cada evento $A\in\mathcal{F}$ un número $P(A)$ satisfaciendo los axiomas:

1. $\forall A \in \mathcal{F}, P(A) \geq 0$
2. Siendo S el espacio muestral, entonces $P(S)=1$
3. Siendo $A_1,A_2,A_3,.... \in \mathcal{F}$ una sucesión de **eventos disjuntos**, $\forall i \neq j, A_i \cap A_j =\emptyset$, entonces: 

$$ P\left [ \displaystyle\bigcup_{i=1}^{\infty} A_i \right ] = \sum_{i=1}^{\infty} P\left [ A_k \right ] $$

Si $S$ fuera un espacio muestral finito el tercer axioma se reduce a una unión finita, y por lo tanto, una suma finita.

Definida la medida de probabilidad, se está en condiciones de estructurar un **espacio de probabilidad** el cuál se nota como $(S,\mathcal{F},P)$, siendo $S$ el espacio muestral, $\mathcal{F}$ la $\sigma$-álgebra y $P$ la medida de probabilidad.

### Complemento.
 
Siendo $A\in\mathcal{F}$ un evento con probabilidad $P(A)$ , su complemento $\overline A$ tiene probabilidad $P (\overline{A} ) = 1-P(A)$.

La demostración es simple. $A\cap \overline A = \emptyset$ y además $A \cup \overline A =S$, luego $P(A)+P( \overline A) =1$, quedando demostrado.

* Siendo $\overline S = \emptyset$ entonces, $P(\emptyset)=0$

## Probabilidad condicional

Se definen dos eventos $A,B \in \mathcal{F}$, se conoce que ocurre $B$, es decir, existe una condición que impone la ocurrencia del evento $B$. Bajo esas condiciones interesa determinar la probabilidad de que ocurra el evento $A$.
Estamos buscando la probabilidad $A$ condicionada por $B$: $P\left ( ^{A}/_{B} \right )$.

En forma intuitiva, si se condiciona a la expresión de $B$, la expresión de $A$ estará dada por la expresión de $A\cap B$, luego podemos estimar la probabilidad condicional:

$$\lim_{n \to \infty} \frac{n_{A\cap B}}{n_B} \rightarrow  P\left ( ^{A}/_{B} \right )$$

siendo $n_{A\cap B}$ y $n_B$ la cantidad de veces que se expresan los eventos $A\cap B$ y $B$ respectivamente. Sea $n$ la cantidad total de experimentos, podemos escribir:


$$\lim_{n \to \infty} \frac{\frac{n_{A\cap B}}{n}}{\frac{n_B}{n}} \rightarrow  \frac{P\left ( {A}\cap{B} \right )}{P\left ({B} \right )}= P\left ( ^{A}/_{B} \right )$$

Luego, $P\left ( {A}\cap{B} \right ) = P\left ({B} \right ) ~ P\left ( ^{A}/_{B} \right )$, o bien, $P\left ( {A}\cap{B} \right ) = P\left ({A} \right ) ~ P\left ( ^{B}/_{A} \right )$.


### Regla de Bayes
Se tiene una partición $B_1, B_2, B_3, ..., B_n$ del espacio muestral $S$, y ocurre un evento $A$, luego
$$ P\left ( ^{B_j}/_{A} \right ) \sum_{i=1} ^n P\left ( ^{A}/_{B_i} \right ) P(B_i)= P\left ( ^{B_j}/_{A} \right ) P(A)=P\left ( ^{A}/_{B_j} \right ) P(B_j)$$

### Eventos Independientes.

Se definen dos eventos $A$ y $B$, se busca determinar la probabilidad de ocurrencia de $A$ condicionado a $B$. Si el evento $B$ no condiciona la ocurrencia del evento $A$, entonces decimos que son independientes, y por lo tanto:

$$P\left ( ^{A}/_{B} \right )=P(A)$$

o en forma equivalente,

$$P\left ( ^{A}/_{B} \right )= \frac{P\left ( {A}\cap{B} \right )}{P\left ({B} \right )}=P(A)$$

por lo tanto, si dos eventos, $A$ y $B$ , son independientes, la probabilidad de que se expresen ambos eventos es igual al producto de las probabilidades de expresión de los eventos:

$$P\left ( {A}\cap{B} \right )=P(B)~P(A)$$

### Quiz

El ejercicio busca estimar el número $\pi$, y para ello se propone estimar el área de un círculo, $\hat A$, de radio $r$, luego, $\hat \pi=\frac{\hat {A}}{r^2}$.

$$ A_1= \left \{ x,y\in [-1,1]~\big /~ \sqrt{x^2+y^2} \leq 1 \right \} $$

Se estima la probabilidad  de que una muestra generada perteneaca a $A_1$:

$$P_{A_1} = \frac{\# (x,y)\in A_1}{\# Total\;de\;muestras\;generadas}$$

La probabilidad $P_{A_1}$ es también la proporción del área del cuadrado que corresponde al círculo.

$$\hat{\pi} = 4 \; P_{A_1}$$

## Variable Aleatoria.


Una variable aleatoria $X$ es una función que asigna un número real $X(\zeta)$, siendo $\zeta \in S$, una muestra del espacio muestral.

Entonces, una variable aleatoria no es más que un mapeo $X:S \rightarrow \mathbb{R}$ (o $\mathbb{C}$)

#### Tipos de variables aleatorias
* **Variable aleatoria discreta**: Es una variable cuyos valores se conforman a partir de un conjunto numerable de eventos. Ese conjunto puede ser o no finito. $X\in\mathbb{N}$, $X\in\{1, 2, 3, 4\}$. Poseen distribuciones discretas de probabilidad:
    * Distribución Bernoulli
    * Distribución Binomial
    * Distribución Poisson
   
* **Variable aleatoria continua**: Es una variable que se distribuye sobre un conjunto contínuo de elementos $X\in[-1,1]$, $X\in\mathbb{R}$, $X\in[0,\infty)$. Poseen distribuciones continuas de probabilidad:
    * Distribución Uniforme
    * Distribución Exponencial
    * Distribución Gaussiana.
    
* **Variable aleatoria mixta**: Es una variable aleatoria que se compone por elementos contínuos y discretos.


### Función distribución acumulativa de probabilidad.

Sea una variable aleatoria contínua $X$, la **función de distribución acumulada** de define como la probabilidad del evento $A=\{a\in S \big / X(a) \leq x \}$:

$$ F_X (x)=P(A)=P\{X\leq x\}$$

#### Propiedades:


* $0\leq F_X (x) \leq 1$
* $\lim_{x \rightarrow \infty} F_X (x)=1$
* $\lim_{x \rightarrow -\infty} F_X (x)=0$
* Siendo $a \leq b$ con $a,b \in \mathbb{R}$ entonces $F_X(a) \leq F_X (b)$
* $F_X (x)$ es continua por derecha
* Se definen $a,b \in \mathbb(R)$, entonces $P(a< X \leq b)= F_X(b)-F_X(a)$
* $P(X>x)= 1- F_X(x)$

### Función densidad de probabilidad

Si existe, la **función densidad de probabilidad** de una variable aleatoria $X$ **continua**, se define como:


$$f_X (x)= \frac {dF_X (x)}{dx}$$

donde, además:

$$ F_X(x) = \int _{-\infty} ^x f_X (x) ~ dx $$

#### Propiedades

* $f_X (x) \geq 0$

* $\int_{-\infty} ^\infty f_X(x)~ dx = 1$

* $P (a\leq X \leq b)= \int_a ^b f_X(x)~ dx$

### Ejemplo: Variable aleatoria Uniforme

Una variable aleatoria uniforme posee la función de distribución acumulada:
\begin{align} 
F_X(x) = \left \{
    \begin{array}\\
        0 &~& x < a \\
        \frac{x-a}{b-a} &~& a \leq x \leq b \\
        1 &~& x>b 
    \end{array} \right.
\end{align}

y su densidad de probabilidad es :

\begin{align} 
f_X(x) = \left \{ 
    \begin{array}\\
        \frac{1}{b-a} &~&  a \leq x \leq b \\
        0 &~& x<a ~y~ x>b
     \end{array} \right.
\end{align}

### Media de una variable aleatoria

Sea $X$ una variable aleatoria, se define la media de $X$, $\mu_X$ como:
$$ \mu_X = E\,[X]=\int_{\{X\}} x \; f_X(x)~ dx  $$

si la variable aleatoria es discreta, entonces se tiene
$$ \mu_X = E\,[X]=\sum_{x \in \{X\}} x \; P(X=x)  $$

#### Estimador de la Media

Sean ${x_i}$ un conjunto de realización de la variable aleatoria $X$, el **promedio** de los $X_i$ es un estimador *insesgado* y *consistente* de la Media de $X$  ($\mu_X$):

$$ \hat{\mu}_X = \frac{1}{N} \sum_{i=1}^N X_i$$

### Media de una función de variable aleatoria.
También se puede plantear la obtener el valor medio, o esperanza que tomará un función $g(x)$ donde $X$ es una variable aleatoria:

$$E\,[g(X)]=\int_{\{X\}} g(x) \; f_X(x)~ dx  $$

si la variable aleatoria es discreta, entonces se tiene

$$E\,[g(X)]=\sum_{x \in \{X\}} g(x) \; P(X=x)  $$

### Varianza de una variable aleatoria.

La varianza es una medida de la dispersión que posee una variable aleatoria:

$$ \sigma^2_X = E \left [(X-\mu_X)^2 \right] = E \left [X^2 \right] -\mu_X^2$$

### Covarianza de variables aleatorias.

Se definen dos variables aleatorias $X_i$ y $X_j$, las cuales poseen medias $\mu_{X_i}$ y $\mu_{X_j}$ respectivamente. Se define a la **covarianza** entre dos variables aleatorias como:

$$\sigma_{X_i\;X_j} = E \left [ (X_i - \mu_{X_i}) (X_j - \mu_{X_j}) \right ] $$

y es una medida de la variabilidad conjunta de dos variables aleatorias.

* Se puede observar que la covarianza de una variable aleatoria consigo misma es la varianza de esa variable aleatoria, $\sigma_{X_i\;X_i} = \sigma_{X_i}^2$.
* Si una variable aleatoria $X_i$ depende linealmente de otra $X_j$, es decir, $\exists \; \alpha \in \mathbb{R} : X_i = \alpha \, X_j \implies \sigma_{X_i\;X_j} = \alpha \; \sigma_{X_j}^2$.

### Desigualdad de Cauchy - Schwarz - Bunyakovsky 

$$|\sigma_{X_i\;X_j}| = \left |E \left [ (X_i - \mu_{X_i}) (X_j - \mu_{X_j}) \right ] \right | \leq \sqrt{\sigma_{X_i}^2\; \sigma_{X_j}^2} $$


* Coeficiente de correlación de Pearsons: Permite estima la **correlación lineal** entre dos variables aleatorias (el test busca rechazar $H_0$ definida como decorrelación entre las variables, y por lo tanto no sirve para demostrar decorrelación):
$$\rho = \frac{\sigma_{X_i\;X_j}}{\sigma_{X_i}\; \sigma_{X_j}} \in [-1,1]$$



## Vectores aleatorios

Se define un conjunto $n$ de variables aleatorias $X_1, X_2, X_3, \cdots,X_n$. En general, estas variables aleatorias pueden responder a diferentes distribuciones de probabilidad. Un vector aleatorio, es simplemente, un arreglo de las variables aleatorias:

$$X = \left [ {\begin{array}\\
X_1 \\
X_2 \\
X_3 \\
\vdots \\
X_n
\end{array}} \right ] $$

Definido un vector aleatorio, se puede determinar una distribución conjunta de probabilidad

$$ \begin{array}\\
F_X(x_1,x_2, \cdots, x_n) & = P(X_1 \leq x_1, X_2 \leq x_2, \cdots, X_n \leq x_n) \\
~& = \int_{-\infty}^{x_1} \int_{-\infty}^{x_2} \cdots \int_{-\infty}^{x_n}  f_{X_1, X_2, \cdots, X_n}  (  x_1, x_2, \cdots, x_n  )~ dx_1\; dx_2\; \cdots dx_n
\end{array} $$

También se pueden definir las densidades de probabilidad marginales:

$$  f_{X_j}(x_j) = \int_{X_1} \cdots \int_{X_{j-1}} \int_{X_{j+1}}\cdots \int_{X_n}  f_{X_1, X_2, \cdots, X_n}  (  x_1, x_2, \cdots, x_n  )~ dx_1\; \cdots dx_{j-1}\; dx_{j+1}\cdots dx_n$$

### Media de un vector aleatorio $\mu_X$.
Siendo $X$ un vector aleatorio, se define a la media del vector $X$ como la esperanza $E[X]$:

$$\mu_X = E[X] = \left [ {\begin{array}\\
E[X_1] \\
E[X_2] \\
E[X_3] \\
\vdots \\
E[X_n]
\end{array}} \right ] $$

### Matriz de Covarianza $\Sigma_X$.

En el caso de un vector aleatorio interesa conocer la **varianza** de cada componente, $\sigma_{X_i}^2 = E[(X_i - \mu_{X_i})^2]$, pero también la relación que existe entre variables aleatorias. Para esto último se puede emplear la **covarianza** permite medir relaciones estadísticas entre dos variables aleatorias:

$$ \sigma (X_i, X_j ) = E_{X_i,X_j} [(X_i - \mu_{X_i})(X_j - \mu_{X_j})] $$

Todas estas medidas estadísticas se contemplan en la **matriz de covarianza** $\Sigma_x$:

$$\Sigma_x = E_x[(X-\mu_x)(X-\mu_x)^T] = \left [ {\begin{array}\\
E\,[(x_1-\mu_{x_1})^2] &  E[(x_1 - \mu_{x_1})(x_2 - \mu_{x_2})] & \cdots & E[(x_1 - \mu_{x_1})(x_n - \mu_{x_n})]\\
E[(x_2 - \mu_{x_2})(x_1 - \mu_{x_1})] & E\,[(x_2-\mu_{x_2})^2] & \cdots & E[(x_1 - \mu_{x_2})(x_n - \mu_{x_n})]\\
\vdots & \vdots & \ddots & \vdots \\
E[(x_n - \mu_{x_n})(x_1 - \mu_{x_1})] &  E[(x_n - \mu_{x_n})(x_2 - \mu_{x_2})] & \cdots & E\,[(x_n-\mu_{x_n})^2]
\end{array}} \right ] $$


#### Propiedades de la Matriz de Covarianza.

Posee propiedades muy importantes para el análisis de procesos aleatorios, 
* Es **simétrica**: $\Sigma_X = \Sigma_X^T$ 
* Ser una matriz simétrica implica ser una matriz normal, y por lo tanto existe una **dioganalización ortogonal**, es decir $$ \forall \Sigma_X \in \mathbb{R}^{nxn}\; t.q.\; \Sigma_X = \Sigma_X ^T \implies \exists P,\Lambda_X \in \mathbb{R}^{nxn}\;t.q. \;  \Sigma_X = P \Lambda_X P^T $$ con $P=P^T$ la matriz de autovectores de $\Sigma_X$ y $\Lambda_X$ una matriz diagonal con los autovalores de  $\Sigma_X$ dispuestos en su diagonal,en un correspondiente con las columnas de $P$. $\Lambda_X$ es en sí misma la matriz de covarianza de la variable $X$ pero expresada en la base de columnas de $P$. Las matrices tiene la forma $$ \begin{array}\\
{P = \left [ {\begin{array}\\
\mid & \mid & \cdots & \mid \\
p_1 & p_2 & \cdots & p_n \\
\mid & \mid & \cdots & \mid
\end{array}} \right ]} & ~~ & {\Lambda_X = \left [ {\begin{array}\\
\lambda_1 & 0 & \cdots & 0 \\
0 & \lambda_2 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & \lambda_n
\end{array}} \right ] } \end{array}$$ siendo $p_i$ autovectores y columnas de $P$ y $\lambda_i$ los autovalores. Hay que notar que al ser $\Sigma$ una matriz simétrica siempre se puede encontrar autovectores ortogonales ($p_i^T p_j = 0$ con $i\neq j$). 

* La matriz de covarianza es **definida no negativa** $\forall z \in \mathbb{R^n}\; t.q.\; z\neq0 : z^T \Sigma_X z \geq 0$
* Los autovalores de la matriz de covarianza son **no negativos**, es decir $\lambda_i \geq 0$. Son en si mismos la varianza de X en la dirección de los autovectores

### Quiz: Vectores aleatorios. Gaussiana Multivariable

Un arreglo n-dimensional de variables aleatorias, $X_1, X_2, X_3,...$, define un vector aleatorio $X = [X_1, X_2, X_3, ..., X_n]^T$. 

Sobre el vector $X$ se pueden definir la media del vector $\mu_x$ y la covarianza del vector $\Sigma_{XX}$:

$$ \mu_x = E[X] $$

$$ \Sigma_{XX} = E[(X-\mu_X)(X-\mu_X)^T]$$

Se pueden dar varios ejemplos de vectores aleatorios. Un vector aleatorio interesante es el formado por variables aleatorias gaussianas. Su función densidad de probabilidad es la siguiente:

$$f_{X_1,\dots,X_n}(x_1,x_2,x_3, ..., x_n) = \frac {1}{(2\pi)^{\frac{n}{2}} |\Sigma|^{\frac{1}{2}}} e ^ {- \frac {1}{2} (X-\mu_X)^T \Sigma ^{-1} (X-\mu_X)}$$

Algunos ejemplos en $\mathbb{R}^2$:

1. $\mu = [0,0]^T ~~~~ \Sigma = \left [ {\begin{array}\\
1 &~& 0  \\
0 &~& 9 
\end{array}} \right ] $ 
2. $\mu = [1,1]^T ~~~~ \Sigma = \left [ {\begin{array}\\
9 &~& 0  \\
0 &~& 1 
\end{array}} \right ] $ 
3. $\mu = [0,0]^T ~~~~ \Sigma = \left [ {\begin{array}\\
1 &~& .999  \\
.999 &~& 1 
\end{array}} \right ] $ 
4. $\mu = [0,0]^T ~~~~ \Sigma = \left [ {\begin{array}\\
1 &~& -1  \\
-1 &~& 4 
\end{array}} \right ] $ 


```python
#caso 1
mu1 = np.asarray([0,0])
sigma1 = np.asarray([[1,0],[0,9]])
vect1 = np.random.multivariate_normal(mu1,sigma1,5000)
#caso 2
mu2 = np.asarray([5,5])
sigma2 = np.asarray([[9,0],[0,1]])
vect2 = np.random.multivariate_normal(mu2,sigma2,5000)
#caso 3
mu3 = np.asarray([0,0])
sigma3 = np.asarray([[1,.999],[.999,1]])
vect3 = np.random.multivariate_normal(mu3,sigma3,5000)
#caso 4
mu4 = np.asarray([0,0])
sigma4 = np.asarray([[1,-1],[-1,4]])
vect4 = np.random.multivariate_normal(mu4,sigma4,5000)

```
